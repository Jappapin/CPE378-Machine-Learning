# Load data

```python
import pandas as pd
cardio_df = pd.read_csv(r'cardio_train_cleaned.csv')
cardio_df
```

```python
cardio_df.info()
```

```python
cardio_result = cardio_df['cardio']
cardio_result
```

```python
del cardio_df['cardio']
```

```python
cardio_feature = cardio_df
cardio_feature
```

```python
cardio_feature.info()
```

```python
# standardizing
from sklearn import preprocessing

std_scaler = preprocessing.StandardScaler()
cardio_feature = pd.DataFrame(std_scaler.fit_transform(cardio_feature), columns = cardio_feature.columns)
cardio_feature
```

# Splitting data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cardio_feature, cardio_result, test_size = 0.3)
```

```python
pd.crosstab(y_train, columns='N', normalize=True)
```

```python
pd.crosstab(y_test, columns='N', normalize=True)
```

# Modeling

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression

Logres_classifier = LogisticRegression()
Logres_classifier.get_params()
```

```python
from sklearn.model_selection import GridSearchCV

params = dict()
params['C'] = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
params["solver"] = ['liblinear', 'newton-cg', 'lbfgs', 'sag', 'saga']
params['max_iter'] = [2500, 5000, 7500, 10000]

gridsearch_ = GridSearchCV(Logres_classifier, params, scoring='accuracy', verbose=2)
gridsearch_.fit(x_train, y_train)
```

```python
print('Best Hyperparameters: %s' % gridsearch_.best_params_)
```

```python
lr = gridsearch_.best_estimator_

acc = round(lr.score(x_train, y_train) * 100, 2)
acc_test = round(lr.score(x_test, y_test) * 100, 2)
print(acc, acc_test)
```

```python
feat_importances = pd.Series(lr.coef_[0], cardio_feature.columns)
feat_importances.nlargest(20).plot(kind='barh',title = 'Feature Importance')
```

### Random Forest

```python
%%time
from sklearn.ensemble import RandomForestClassifier
Randfor_classifier = RandomForestClassifier()
Randfor_classifier.get_params()
```

```python
params = dict()
params['n_estimators'] = [100, 200, 300]
params['max_depth'] = [2, 4, 8, 16, None]
params["max_features"] = ['auto', 'sqrt']
params["min_samples_leaf"] = [1, 2, 3]
params["min_samples_split"] = [2, 5, 10]
params["criterion"] = ["gini", "entropy"]
params['class_weight'] = ['balanced','balanced_subsample']
params['bootstrap'] = [True]
```

```python
gridsearch = GridSearchCV(Randfor_classifier, params, scoring='accuracy', verbose=3)
gridsearch.fit(x_train, y_train)
```

```python
print('Best Hyperparameters: %s' % gridsearch.best_params_)
```

```python
rf = gridsearch.best_estimator_

acc = round(rf.score(x_train, y_train) * 100, 2)
acc_test = round(rf.score(x_test, y_test) * 100, 2)
print(acc, acc_test)
```

```python
import matplotlib.pylab as plt
feat_imp = pd.Series(rf.feature_importances_, cardio_feature.columns).sort_values(ascending=False)
feat_imp.plot(kind='barh', title='Feature Importances')
plt.ylabel('Feature Importance Score')
```

## Evaluation

```python
from sklearn.metrics import classification_report
```

```python
lr_result = lr.predict(x_test)
print(classification_report(y_true=y_test, y_pred=lr_result))
```

```python
rf_result = rf.predict(x_test)
print(classification_report(y_true=y_test, y_pred=rf_result))
```

```python
from sklearn.datasets import make_classification
x_synthetic, y_synthetic = make_classification(n_samples= 60000, 
                                               n_features=10, 
                                               shuffle=True, 
                                               random_state=42)
```

```python
x_synthetic
```

```python
y_synthetic
```

```python
# split train and test set
x_synthetic_train, x_synthetic_test, y_synthetic_train, y_synthetic_test = train_test_split(x_synthetic, 
                                                                                            y_synthetic, test_size = 0.3)
```

```python
lr_syn = LogisticRegression(C=0.1, max_iter=2500, solver='liblinear')
lr_syn.fit(x_synthetic_train, y_synthetic_train)
```

```python
acc = round(lr_syn.score(x_synthetic_train, y_synthetic_train) * 100, 2)
acc_test = round(lr_syn.score(x_synthetic_test, y_synthetic_test) * 100, 2)
print(acc, acc_test)
```

```python
rf_syn = RandomForestClassifier(bootstrap=True, class_weight='balanced', 
                                criterion='gini', max_depth=8, 
                                max_features='sqrt', min_samples_leaf=3, 
                                min_samples_split=2, n_estimators=200)
rf_syn.fit(x_synthetic_train, y_synthetic_train)
```

```python
acc = round(rf_syn.score(x_synthetic_train, y_synthetic_train) * 100, 2)
acc_test = round(rf_syn.score(x_synthetic_test, y_synthetic_test) * 100, 2)
print(acc, acc_test)
```

```python
from xgboost import XGBClassifier
import xgboost
```

```python
xgb_syn = xgboost.XGBClassifier(colsample_bytree = 0.5, gamma = 1,
                                  learning_rate = 0.01, min_child_weight = 1,
                                  max_depth = 5, scale_pos_weight = 1, n_estimators=100)
xgb_syn.fit(x_synthetic_train, y_synthetic_train)
```

```python
acc = round(xgb_syn.score(x_synthetic_train, y_synthetic_train) * 100, 2)
acc_test = round(xgb_syn.score(x_synthetic_test, y_synthetic_test) * 100, 2)
print(acc, acc_test)
```
